Task 1:Load and Inspect the Data

```python
import pandas as pd
import numpy as np
data = {
    "transaction_id":range(1,21),
    "data":pd.date_range('2024-10-01', periods =20,freq ='D'),
    "region":['North','South','East','West',None,'North','South',None,'East','West',
              'North','South','East','West','North',None,'East','West','North','South'],
    "Product_category":['Electronics','Clothing',None,'Books','Electronics','Home','Clothing',
                        'Books','Electronics',None,'Home','Clothing','Books','Electronics',
                        'Home','Clothing','Books','Electronics','Home','Clothing'],
    "sales_amount":[1200,450,890,None,1500,670,None,340,2100,780,560,None,420,1800,
                    920,510,380,None,1100,640],
    "quantity":[2,5,3,1,None,4,2,3,1,5,3,2,None,1,4,3,2,1,None,4],
    "customer_age":[25,34,None,45,29,None,38,52,27,41,
                    33,None,48,26,35,42,None,31,39,44],
    "payment_method":['Credit Card','UPI','Cash','Debit Card','Credit Card',
                      'UPI','Cash',None,'Credit Card','UPI','Debit Card',
                      'Cash','Credit Card',None,'UPI','Cash','Debit Card',
                      'Credit Card','UPI',None]
}
df = pd.DataFrame(data)
df
```

Task 2:Handle Missing Values

1:Region & Product_category:Fill with mode(most frequent value)

```python
mode_region = df['region'].mode()[0]
df['region']= df['region'].fillna(mode_region)
```

```python
mode_product=df['Product_category'].mode()[0][1]
df['Product_category']=df['Product_category'].fillna(mode_product)
```

2:Sales_amount: Fill with median

```python
median_sales = df['sales_amount'].median()
df['sales_amount'] = df['sales_amount'].fillna(median_sales)
```

3:Quantity :Fill using forward fill (ffill)

```python
df['quantity']=df['quantity'].fillna(method='ffill')
```

4.Customer_age:Fill with mean(rounded to integer)

```python
mean_customer = df['customer_age'].mean()
df['customer_age']=df['customer_age'].fillna(round(mean_customer))
mean_customer
```

5.Payment_method: Drop rows where payment_method is missing

```python
df = df.dropna()
df
```

Verify no missing values remain using df.isna().sum()

```python
df.isna().sum()
```

Task : 4 Custom Aggregation

1.Create a function sales_range() that returns max - min of sales

```python
def sales_range(x):
    return x.max()-x.min()
sales_range =df['sales_amount'].agg(sales_range)
sales_range
```

2.Apply it to find sales range for each region

```python
def sales_range(x):
    return x.max()-x.min()
range = df.groupby('region')['sales_amount'].agg(sales_range)
range
```

3 . use.agg() to calculate multilple metrics by region:


    . sales_amount : sum,mean,max
    . quantity : sum,min

```python
result =df.groupby('region').agg({"sales_amount":["sum","mean","max"],
                                 "quantity":["sum","min"] })
result
```
